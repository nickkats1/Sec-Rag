```python
from src.pipeline.data_ingestion import DataIngestion
from src.pipeline.vector_store import VectorStore
from src.pipeline.rag import Rag

from IPython.display import Markdown

# load in instances of DataIngestion, VectorStore and Rag
data_ingestion = DataIngestion(file_path="data/")
vs = VectorStore(file_path="data/")
rag = Rag(model="gpt-4", temperature=0)

# documents and chunks from DataIngestion class
documents = data_ingestion.load_documents()
chunks = data_ingestion.chunk(chunk_size=2500, chunk_overlap=250)


# vector store and retriever from VectorStore
vector_store = vs.load_vector_store(persist_directory="./db", chunks=chunks)
retriever = vs.load_retriever(vector_store, search_type="similarity", k=8)

# result from Rag class
result = rag.prompt_rag(company_name="meta", retriever=retriever)
Markdown(result)
```
