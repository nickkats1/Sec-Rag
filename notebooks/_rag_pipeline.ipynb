## Example Usage.

```python
import os

from langchain_core.prompts import PromptTemplate
from rich.markdown import Markdown
from rag.data_ingestion import load_documents, chunk_documents
from rag.llm import LLM
from rag.retriever import get_retriever

import warnings
warnings.filterwarnings("ignore")

llm = LLM(
    api_key=os.getenv("GROQ_API_KEY")).get_llm(provider="groq", model_name="llama-3.3-70b-versatile", temperature=0.0
    )


file_path = "google_10K.pdf"

# load documents
docs = load_documents(file_path)
chunks = chunk_documents(docs, chunk_size=2010, chunk_overlap=200)

retriever = get_retriever(chunks, k = 8, search_type="similarity")

question = input("Enter your question: ")
context = retriever.invoke(question)

prompt = PromptTemplate(
    input_variables=['question', 'context'],
    template="""Answer the question given the provided context the best you can. If you do not know the answer, just say you do not know.
    Do not make anything up. Only answer the question and do not add anything extra.
    
    Context:
    {context}
    
    Question:
    {question}
    Answer"""
)

chain = prompt | llm

result = chain.invoke({"question": question, "context": context}).content

Markdown(result)
```
