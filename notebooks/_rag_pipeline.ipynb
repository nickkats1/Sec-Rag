- Example usage

```python
from src.rag_pipeline import (
    load_documents,
    chunk_documents,
    build_vector_store,
    build_retriever,
    run_rag
)

# load documents from file path
documents = load_documents("data/google_10K.pdf")
```

```python
# chunk documents and give chunk size and chunk overlap
chunks = chunk_documents(
    documents=documents,
    chunk_size=1000,
    chunk_overlap=50
)
```

```python
# vector store
vector_store = build_vector_store(
    chunks=chunks,
    persist_directory="./db"
)

# retriever using chromadb
retriever = build_retriever(
    vector_store=vector_store,
    search_type="similarity",
    k=8
)
```

```python
# ask question
question = "How much did Google spend on Research and development in 2025 and previous years?"
run_rag(question, retriever,model="gpt-3.5-turbo",temperature=0)
```
