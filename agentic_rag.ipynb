```python

def summary_node(state: State) -> State:
    """State for Agent"""
    prompt = PromptTemplate(
        input_variables=["documents"],
        template=SummarizationPrompts
    )
    message = HumanMessage(content=prompt.format(documents=state['documents']))
    summary = llm.invoke([message]).content
    return {"summary":summary}


def balance_sheet_node(state: State) -> State:
    """Node balance sheets"""
    prompt = PromptTemplate(
        input_variables=["documents"],
        template=BalanceSheetPrompts
    )
    message = HumanMessage(content=prompt.format(documents=state['documents']))
    balance_sheet = llm.invoke([message])
    return {"balance_sheet": balance_sheet}
```
